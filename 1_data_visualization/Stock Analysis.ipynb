# **1.2 Lưu dữ liệu từ 5 ngân hàng TCB, ACB, VCB, CTG, MBB**

```python
#Chọn thời gian lấy dữ liệu
start_date_selected="2020-01-01"
end_date_selected='2023-03-19'

#Lấy dữ liệu từ 5 ngân hàng và lưu vào dataframe
TCB =  stock_historical_data(symbol='TCB', start_date=start_date_selected, end_date=end_date_selected)
ACB =  stock_historical_data(symbol='ACB', start_date=start_date_selected, end_date=end_date_selected)
VCB =  stock_historical_data(symbol='VCB', start_date=start_date_selected, end_date=end_date_selected)
CTG =  stock_historical_data(symbol='CTG', start_date=start_date_selected, end_date=end_date_selected)
MBB =  stock_historical_data(symbol='MBB', start_date=start_date_selected, end_date=end_date_selected)
#Đặt index column theo Trading Date
TCB = TCB.set_index('TradingDate')
ACB = ACB.set_index('TradingDate')
VCB = VCB.set_index('TradingDate')
CTG = CTG.set_index('TradingDate')
MBB = MBB.set_index('TradingDate')
TCB.tail() #lấy thông tin 5 dòng cuối
```

```python
#Để nối các dữ liệu trên ta dùng phương thức concat() và truyền vào 1 list là các data frame trên
#Trước tiên ta tạo 1 list_banks là tên các mã cổ phiếu trên
list_banks = ['TCB', 'ACB', 'VCB', 'CTG', 'MBB']
# Tham số axis = 1 là nối theo cột
bank_stocks = pd.concat([TCB, ACB, VCB, CTG, MBB],axis=1, keys = list_banks)
bank_stocks
```

```python
#Tiếp theo đặt tên cột cho data frame trên
bank_stocks.columns.names = ['Bank Name','Stock Info']
bank_stocks
```

```python
#Sử dụng phương thức .xs() trong pandas để lập chỉ mục đa cấp để tìm giá trị đóng cửa lớn nhất của từng cổ phiếu trên
```

## **1.3 Tính tỉ suất lợi nhuận**

```python
#Giờ ta tính tỉ suất lợi nhuận
value_banks = pd.DataFrame()
for name in list_banks:
    value_banks[name+' value_bank'] = bank_stocks[name]['Close'].pct_change()
value_banks.dropna(inplace= True)#Vì dòng đầu tiên không có dữ liệu để tính nên trả về giá trị NaN, ta có thể xóa luôn dòng này
value_banks.head()
```

## **2.1 Trực quan hoá dữ liệu**

*Nhận* xét: Thông qua biểu đồ pairplot trên, có thể phát hiện được mối tương quan giữa các cặp biến số trong dữ liệu.
```python
#Vẽ đường TCB 7, 14, 21 cổ phiếu VCB
#Đường SMA là gì? Đường Simple Moving Average (SMA): là đường trung bình động đơn giản được tính bằng trung bình cộng các mức giá đóng cửa trong một khoảng thời gian giao dịch nhất định.
VCB['Close'].loc['2022-01-01':'2023-01-01'].ta_plot(study='sma',periods=[7,14,21],title='SMA TCB')
```

```python
#Vẽ dãi bollinger đơn giản cổ phiếu VCB trong năm 2022
VCB['Close'].loc['2022-01-01':'2023-01-01'].ta_plot(study='boll')
```

Nhận xét: Với dãi Bollinger, ngưởi dùng có thể dễ dàng nhìn thấy sự biến động giá của cổ phiếu, tìm ra điểm bắt đầu của cổ phiếu trong xu hướng tăng/giảm giá và đưa ra quyết định mua/bán cổ phiếu dựa trên sự chuyển động của giá cổ phiếu. Khi dãi Bollinger co lại hoặc mở ra, điều này cũng thể hiện tính chất của quá trình giao dịch trên thị trường. Nếu băng Bollinger co lại, có thể xuất hiện tín hiệu giá sắp đứng ở mức cực độ hàng hoặc giá đã tạo đáy và sẽ khái niệm trở lại giá cổ phiếu. Nếu băng đưa ra, thì các nhà giao dịch có thể đánh giá giá cổ phiếu có xu hướng bứt phá hay lùi lại trong tương lai.
